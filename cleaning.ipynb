```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Convert 'TotalCharges' to numeric, coerce errors to NaN
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')

# Fill missing values in 'TotalCharges' with the median
data['TotalCharges'].fillna(data['TotalCharges'].median(), inplace=True)

# Convert 'Churn' to binary (0 = No, 1 = Yes)
data['Churn'] = data['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

# Convert 'SeniorCitizen' to binary (0 = No, 1 = Yes)
data['SeniorCitizen'] = data['SeniorCitizen'].astype(int)

# Display the first few rows to check the changes
data.head()
```

```python
# Create a 'tenure_group' column
bins = [0, 12, 24, 36, 48, 60, np.inf]
labels = ['0-12', '12-24', '24-36', '36-48', '48-60', '60+']
data['tenure_group'] = pd.cut(data['tenure'], bins=bins, labels=labels, right=False)

# Create a 'monthly_charges_group' column
bins = [0, 35, 70, 100, np.inf]
labels = ['0-35', '35-70', '70-100', '100+']
data['monthly_charges_group'] = pd.cut(data['MonthlyCharges'], bins=bins, labels=labels, right=False)

# Display the first few rows to check the new columns
data.head()
```

```python
data.to_csv('cleaned_telco_customer_churn.csv',index=False)
```
